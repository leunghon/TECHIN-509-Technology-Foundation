```python
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 3 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
for i in range(1,106):
    if i % 3 == 0 and i % 5 == 0 and i % 7 == 0:
        print('FizzBuzzQizz')
    elif i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
    elif i % 3 == 0 and i % 7 == 0:
        print('FizzQizz')
    elif i % 5 == 0 and i % 7 == 0:
        print('BuzzQizz')
    elif i % 7 == 0:
        print('Qizz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 3 == 0:
        print("Buzz")
    else:
        print(i)
```
